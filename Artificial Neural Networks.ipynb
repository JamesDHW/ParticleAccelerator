## The Neural Network

#### The Code

This neural network attempts to classify events to one of two outcomes (male or female) based on two arbitrary input variables (as shown on the graph axes). The network has three layers - an input, an output and one hidden layer of alterable size. The input is given by X,y = sklearn.datasets.make_moons(200, noise=0.60), which uses np.random.seed(0) to create a pseudorandom distribution in the shape of two interlocking moons of male and female outputs. Here, X gives input coordinates of the two variables and y gives the output of the corresponding event (1 or 0 for male or female).

The neural network has 2 nodes in the first layer, which are the variables in X. The second (hidden layer) has a changing number of nodes for each exercise. The third layer has two nodes again (two outputs - male/ female).

#### Changing the Hidden Layer Size

Changing the input of the build_model function: build_model(hidden_layer_size, print_loss=True)

As the hidden layer size increases a more and more precise guess for the output can be made. This is seen when we compare the straighter line shape of a hidden layer size 1 compared to the more twisted/ jagged shape for a hidden layer size of 50. Ideally the classification should be sharp, but should not try to over-classify events as this strays into the relm of simply selecting where each male or female outcome event occurs in the training data. This is not useful as when you run actual data through the network (as the input of events are slightly randomised) the classification won't be useful as individual event points will have changed. This is to say that the neural network should look for a trend rather than picking out individual data points, which will occur if the network is over-trained. In this case, from observation, a hiden layer size of 4 seems to be appropriate (that is 4 nodes in the hidden layer).

#### Using More Events

Changing the input of the number of events in the generator: sklearn.datasets.make_moons(number_of_events, noise=0.60)

Increasing the number of events gives the NN more data to train its output guesses. This means that increasing the number of events makes for more accurate output guesses, as seen by the higher proportion of correct guesses when we compare 1000 events to 500 events (a much higher percentage of events are correctly classified when given 1000 events). This shows that maximising the number of trining events is always beneficial (unlike maximising the number of hidden layers), however in practice this is not always possible as there may only be a certain amount of training data avaiable.

#### Increasing the Noise

Changing the input of the noise in the generator: sklearn.datasets.make_moons(200, noise)

Increasing the amount of noise decreased the strength of the interlocking moons pattern which means that the neural network has a more difficult pattern to find. To the untrained eye there is very little pattern when the noise is 0.6. This would need many more events to start to see a pattern thereofre the fitting is poor.
